```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# DESCRIPTIVE ANALYTICS FOR NUMERICAL COLUMNS

# Reading dataset
df = pd.read_csv("sales_data_with_discounts.csv")
```

```python
# Identifying numerical columns
numerical_columns = df.select_dtypes(include=['int64','float64']).columns
```

```python
numerical_columns
```

```python
# Calculating Basic Statistics
```

```python
# Mean
mean = df[numerical_columns].mean()
print("mean:\n",mean)
```

```python
# Median
median = df[numerical_columns].median()
print("median:\n",median)
```

```python
# Mode
mode = df[numerical_columns].mode().iloc[0]
print("mode:\n",mode)
```

```python
# Standard Deviation
std = df[numerical_columns].std()
print("standard deviation:\n",std)
```

```python
# DATA VISUALIZATION

numerical_columns
```

```python
# Histograms for numerical columns
for column in numerical_columns:
    plt.figure(figsize=(10,8))
    plt.hist(df[column], bins=10, color='blue',edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Creating boxplots for numerical columns
plt.figure(figsize=(10,8))
df[numerical_columns].boxplot()
plt.title('Boxplots of Numerical Columns')
plt.ylabel('Value')
plt.show()
```

```python
# Identifying categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
categorical_columns
```

```python
# Creating Bar charts for categorical columns
for column in categorical_columns:
    plt.figure(figsize=(10,8))
    df[column].value_counts().plot(kind='bar',color='skyblue')
    plt.title(f'Bar Chart of{column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# STANDARDIZATION OF NUMERICAL VARIABLES

def standardize_column(column):
   mean = column.mean()
   std = column.std()
   standardized_column = (column - mean) / std
   return standardized_column
```

```python
# Applying standardization
for column in numerical_columns:
       df[column] = standardize_column(df[column])

print(df[numerical_columns].head())
```

```python
# CONVERSION OF CATEGORICAL DATA INTO DUMMY VARIABLES

dummy_df = pd.get_dummies(df[categorical_columns])
dummy_df = dummy_df.astype(int)
print("Dummy Variables:\n",dummy_df.head())
```

```python
# combining numerical data with dummy variables
df_with_dummies = pd.concat([df.drop(categorical_columns,axis=1),dummy_df],axis=1)
print(df_with_dummies.head())
```
